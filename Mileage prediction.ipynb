Define Target variable y and feature x

```python
df.columns
```

```python
y=df['mpg']
```

```python
y.shape
```

```python
X=df[['displacement','horsepower','weight','acceleration']]
```

```python
X.shape
```

```python
X
```

Scaling data

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
X=ss.fit_transform(X)
```

```python
X
```

```python
pd. DataFrame(X). describe ()
```

Train test spilit data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.7,random_state=2529)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

Linear regression model

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression ()
```

```python
lr.fit(X_train,y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

Mileage=23.4-1.05Displacement-1.68Horsepower-4.10Weight-0.115Acceleration+error

Predict Test Data

```python
y_pred=lr.predict(X_test)
```

```python
y_pred
```

Model Accuracy

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,r2_score
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly=PolynomialFeatures(degree=2, interaction_only=True,include_bias=False)
```

```python
X_train2=poly.fit_transform(X_train)
```

```python
X_test2=poly.fit_transform(X_test)
```

```python
lr.fit(X_train2,y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
y_pred_poly=lr.predict(X_test2)
```

Model Accuracy

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_error,mean_absolute_percentage_error,r2_score
```

```python
mean_absolute_error(y_test,y_pred_poly)
```
